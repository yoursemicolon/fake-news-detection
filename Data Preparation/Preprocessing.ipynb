# **Text Preprocessing**

Pada langkah ini, akan dilakukan:
1. Case Folding, yaitu mengkonversi text menjadi bentuk standar. Membuat semua huruf menjadi lowercase.
2. Tokenizing berupa penghapusan white space dan tanda baca
3. Filtering berupa stopwords removal
4. Lemmatization, yaitu proses mengubah sebuah kata menjadi benuk yang sesuai (lemma) sehingga dapat dikelompokkan dengan kata lain yang sama.

## **Import Library**

```python
import pandas as pd
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# checking the columns of dataset
features = pd.read_csv("../Dataset/Construction Data/features.csv")
print(list(features))
```

## **Case Folding**

```python
# converting all words to lowercase
features['title'] = features['title'].str.lower()
features['text'] = features['text'].str.lower()
features
```

## **Tokenizing**

### **Whitecase and Punctuation Removal**

Kita menggunakan method word_tokenize() untuk memisahkan teks menjadi potongan-potongan kata. Pada tahap ini, dilakukan proses whitecase removal dan punctuation removal.

```python
# remove whitespace and punctuation

def identify_title_tokens(row):
    title = row['title']
    tokens_for_title = nltk.word_tokenize(title)
    # taken only words and numbers (not punctuation)
    token_words_for_title = [w for w in tokens_for_title if w.isalnum()]
    return token_words_for_title

def identify_text_tokens(row):
    text = row['text']
    tokens_for_text = nltk.word_tokenize(text)
    # taken only words and numbers (not punctuation)
    token_words_for_text = [w for w in tokens_for_text if w.isalnum()]
    return token_words_for_text
```

Menggunakan fungsi .apply() milik Pandas DataFrame untuk menjalankan function external pada kolom. Hasil word_tokenize() disimpan pada kolom terpisah dengan nama title_tokenized dan text_tokenized.

```python
# make new columns in dataframe with all tokenized words

features['title_tokenized'] = features.apply(identify_title_tokens, axis=1)
features['text_tokenized'] = features.apply(identify_text_tokens, axis=1)
features
```

## **Filtering/Stopwords Removal**

Pada proses ini akan dilakukan stopwords removal. Filtering ini bertujuan untuk mengambil kata-kata penting pada tokens yang dihasilkan pada proses sebelumnya. **Stopwords** adalah kata-kata umum yang biasanya muncul dalam jumlah besar dan dianggap tidak memiliki makna. Contoh stopwords dalam bahasa Inggris adalah "a", "the", "is", "are", dan lain-lain. 

Kita menggunakan stopword bahasa Inggris yang didapatkan dari library NLTK untuk filtering terhadap DataFrame. Penggunaan fungsi **.set()** berguna untuk membuat iterable list menjadi sequence iterable element. Hasilnya kita dapat mempercepat proses pengecekan apakan sebuah token terdapat pada **list_stopword** atau tidak **(if *token* not in *list_stopword*)**.



```python
from nltk.corpus import stopwords
list_stopword = set(stopwords.words("english"))

def remove_title_stops(row):
    words = row["title_tokenized"]
    meaningful_words = [w for w in words if w not in list_stopword]
    return meaningful_words

def remove_text_stops(row):
    words = row["text_tokenized"]
    meaningful_words = [w for w in words if w not in list_stopword]
    return meaningful_words
    
```

Menggunakan fungsi .apply() milik Pandas DataFrame untuk menjalankan function external pada kolom. Hasil filtering disimpan pada kolom terpisah dengan nama title_no_stopwords dan text_no_stopwords.

```python
# make new columns in dataframe with words after stopwords removal

features['title_no_stopwords'] = features.apply(remove_title_stops, axis=1)
features['text_no_stopwords'] = features.apply(remove_text_stops, axis=1)
features
```

## **Lemmatization**

Lematisasi berusaha untuk mengubah token atau term menjadi kata dasarnya. Tujuan dari lematisasi adalah mengubah infinite tense dan noun menjadi sebuah kata dalam Bahasa Inggris yang sama. Pada proses ini, kita akan menggunakan **WordNet Lemmatizer algorithm**.

```python
# lemmatizing words

from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 

def lemmatize_title_list(row):
    list = row['title_no_stopwords']
    lemmatized_list = [lemmatizer.lemmatize(word) for word in list]
    return lemmatized_list

def lemmatize_text_list(row):
    list = row['text_no_stopwords']
    lemmatized_list = [lemmatizer.lemmatize(word) for word in list]
    return lemmatized_list
```

```python
# make new columns in dataframe with lemmatized lists

features['lemmatized_title'] = features.apply(lemmatize_title_list, axis=1)
features['lemmatized_text'] = features.apply(lemmatize_text_list, axis=1)
features
```

## **Rejoining**

Pada proses ini, kita menyatukan kembali array tokens pada lemmatized_title dan lemmatized_word menjadi satu kalimat utuh.

```python
# Rejoining words

def rejoin_title_words(row):
    list = row['lemmatized_title']
    joined_words = ( " ".join(list))
    return joined_words

def rejoin_text_words(row):
    list = row['lemmatized_text']
    joined_words = ( " ".join(list))
    return joined_words
```

```python
# make new columns in dataframe with joined words

features['processed_title'] = features.apply(rejoin_title_words, axis=1)
features['processed_text'] = features.apply(rejoin_text_words, axis=1)
features
```

## **Feature Selection (again)**

Drop semua kolom yang ada kecuali processed_title dan processed_text

```python
# drop columns

columns = ['title', 'text', 'title_tokenized', 'text_tokenized', 'title_no_stopwords', 
                'text_no_stopwords', 'lemmatized_title', 'lemmatized_text']
features.drop(columns, axis=1, inplace=True)
features
```

## **CountVectorizer**

**CountVectorizer** dapat menghitung frekuensi kata dan mengubah fitur teks menjadi sebuah representasi vektor (encode) untuk digunakan sebagai input dalam algoritme pembelajaran mesin. Pada proses ini, kita akan menggunakan CountVectorizer untuk menghitung frekuensi pada processed_title dan processed_text. Parameter yang digunakan adalah ngram_range, yaitu potongan n-token yang diambil dari string/kalimat. ngram_range diset menjadi (2,2) untuk token dua-dua karakter. ngram_range bahu-membahu dengan analyzer sebagai parameter kedua. Analyzer diset menjadi "Word" agar menghasilkan kata atau frasa. 

Selain itu, digunakan juga method fit_transform() yang merupakan penggabungan dari dua method: fit() dan transform() api. fit_transform() method melakukan kalkulasi dan transformasi.

```python
from sklearn.feature_extraction.text import CountVectorizer
from scipy import sparse

word_vectorizer1 = CountVectorizer(ngram_range=(2, 2), analyzer='word')
sparse_matrix_for_title = word_vectorizer1.fit_transform(features['processed_title'])
sparse.save_npz("sparse_matrix_for_title.npz", sparse_matrix_for_title)
sparse_matrix_for_title = sparse.load_npz("sparse_matrix_for_title.npz")
```

```python
word_vectorizer2 = CountVectorizer(ngram_range=(2, 2), analyzer='word')
sparse_matrix_for_text = word_vectorizer2.fit_transform(features['processed_text'])
sparse.save_npz("sparse_matrix_for_text.npz", sparse_matrix_for_text)
sparse_matrix_for_text = sparse.load_npz("sparse_matrix_for_text.npz")
```
